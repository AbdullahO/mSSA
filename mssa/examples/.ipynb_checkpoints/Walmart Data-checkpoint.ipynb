```python
# Import the packages
from mssa.mssa import mSSA
import pandas as pd
import matplotlib.pyplot as plt
import io
import numpy as np

```

```python
data = pd.read_csv("../../../../../Downloads/walmart-recruiting-store-sales-forecasting 2/train.csv")
data
```

```python
ts = data.pivot_table(index=['Store','Dept'], columns=['Date'], values='Weekly_Sales').T
ts.columns = ['%s_%s'%(i,j) for (i,j) in ts.columns ]
ts = ts.reset_index()
ts.index = pd.to_datetime(ts.Date)
ts = ts.iloc[:,1:]
ts
```

```python
model = mSSA(fill_in_missing = True, col_to_row_ratio = 3, rank = 15)
# model
model.update_model(ts.iloc[:-20,:])
```

```python
model.ts_model.models[0].matrix.shape
model.k
```

```python
ts.index[-20]
ts_hat = np.zeros(ts.values[-20:,:].shape)
for i,col in enumerate(ts.columns):
    df = model.predict(col,ts.index[-20],ts.index[-1])
    ts_hat[:,i] = df['Mean Predictions'].values
# ts.iloc[:-20,:]
plt.plot(ts.values[-20:,0])
plt.plot(ts_hat[:,0])
np.nanmean(np.abs(ts.values[-20:,:]-ts_hat))
```

```python
np.nanmean(np.abs(ts.values[-20:,:]-ts_hat))
```
